```python
'''3.Explain the following terms:
> The difference in npYarray, npYasarray and npYasanyarrayX
> The difference between Deep copy and shallow copy'''

'''In NumPy, np.array, np.asarray, and np.asanyarray are functions used to create and convert arrays, but they have subtle differences in how they handle the input data types and behaviors. Here's a detailed explanation of each:

1. np.array
Functionality: This function creates a new NumPy array from a given object (like a list, tuple, or another array).

Behavior:

It always returns a new array, even if the input is already a NumPy array.
The new array is created with the specified dtype (data type) unless otherwise specified.
It does not preserve the type of subclasses. For example, if you pass a NumPy masked array or a subclass of ndarray, it will return a standard ndarray.
Example:

import numpy as np

# Creating a NumPy array
arr = np.array([1, 2, 3])
print(arr)  # Output: [1 2 3]
print(type(arr))  # Output: <class 'numpy.ndarray'>

2. np.asarray
Functionality: This function converts the input to an array but is more flexible regarding the input type.

Behavior:

It returns the input array itself if the input is already an ndarray. This means it does not create a copy of the array if it's already a NumPy array, which can be more efficient in terms of memory and performance.
It also converts other objects (like lists, tuples) into an array, similarly to np.array, but with the notable difference of not copying existing arrays.
If the input is a subclass of ndarray, it preserves the subclass type.
Example:


import numpy as np

# Existing NumPy array
existing_arr = np.array([1, 2, 3])

# Using np.asarray on the existing array
arr = np.asarray(existing_arr)
print(arr is existing_arr)  # Output: True (no new array is created)

# Using np.asarray on a list
arr2 = np.asarray([4, 5, 6])
print(arr2)  # Output: [4 5 6]

3. np.asanyarray
Functionality: This function behaves similarly to np.asarray, but it allows for more flexibility in terms of the input type.

Behavior:

It preserves subclasses of ndarray (like masked arrays or matrix) and returns them as is.
Like np.asarray, it does not create a new array if the input is already an ndarray.
Example:


import numpy as np

# Using np.asanyarray with a regular NumPy array
arr = np.array([1, 2, 3])
result = np.asanyarray(arr)
print(result is arr)  # Output: True (no new array is created)

# Using np.asanyarray with a masked array
masked_arr = np.ma.array([1, 2, 3], mask=[0, 1, 0])
result2 = np.asanyarray(masked_arr)
print(type(result2))  # Output: <class 'numpy.ma.core.MaskedArray'>
```

```python
''' 5.Create a NumPy array with random integers between 1 and 100 of shape (5,6). After creating the array
perform the following operations:

 a)Extract all even integers from array.

 b)Extract all odd integers from array'''

import numpy as np

# Create a 5x6 array with random integers between 1 and 100
random_array = np.random.randint(1, 101, size=(5, 6))

# Output the original array
print("Original 5x6 Array with Random Integers:")
print(random_array)

# a) Extract all even integers
even_integers = random_array[random_array % 2 == 0]
print("\nEven Integers:")
print(even_integers)

# b) Extract all odd integers
odd_integers = random_array[random_array % 2 != 0]
print("\nOdd Integers:")
print(odd_integers)
```

```python
'''14.Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity'). 

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a time-series dataset
date_range = pd.date_range(start='2024-01-01', periods=30, freq='D')  # Generate 30 daily dates
temperature = np.random.uniform(15, 35, size=30)  # Random temperatures between 15 and 35 degrees Celsius
humidity = np.random.uniform(40, 90, size=30)     # Random humidity levels between 40% and 90%

# Create the DataFrame
df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Step 2: Plot 'Temperature' and 'Humidity' on the same plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(10,6))

# Plot 'Temperature' on the left y-axis
ax1.plot(df['Date'], df['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')  # b) Label the x-axis as 'Date'
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for 'Humidity'
ax2 = ax1.twinx()  # Instantiate a second axes that shares the same x-axis
ax2.plot(df['Date'], df['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Step 3: Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

```python
'''15.Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following  tasks using Matplotlib:
a) Plot a histogram of the data with 30 bins. 

b) Overlay a line plot representing the normal distribution's probability density function (PDF). 

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean = 0, Std dev = 1

# Step 2: Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Step 3: Overlay the PDF of the normal distribution
xmin, xmax = plt.xlim()  # Get the current x-axis limits
x = np.linspace(xmin, xmax, 100)  # Create a range of x values for the PDF
p = norm.pdf(x, np.mean(data), np.std(data))  # Compute the PDF with the same mean and std deviation as the data
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Step 4: Label the x-axis and y-axis
plt.xlabel('Value')  # c) Label x-axis as 'Value'
plt.ylabel('Frequency/Probability')  # c) Label y-axis as 'Frequency/Probability'

# Step 5: Set the title of the plot
plt.title('Histogram with PDF Overlay')  # d) Set the title of the plot

# Step 6: Display a legend
plt.legend()

# Step 7: Show the plot
plt.show()
```

```python
'''17.Create a Seaborn scatter plot of two random arrays, 
      color points based on their position relative to the origin (quadrants),
      add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.'''


import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate two random arrays for x and y
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)  # 100 random values for x
y = np.random.randn(100)  # 100 random values for y

# Step 2: Determine the quadrant based on the sign of x and y
def determine_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Q1'  # First quadrant
    elif x < 0 and y >= 0:
        return 'Q2'  # Second quadrant
    elif x < 0 and y < 0:
        return 'Q3'  # Third quadrant
    else:
        return 'Q4'  # Fourth quadrant

# Create a DataFrame with x, y, and their corresponding quadrants
df = pd.DataFrame({'x': x, 'y': y})
df['Quadrant'] = df.apply(lambda row: determine_quadrant(row['x'], row['y']), axis=1)

# Step 3: Create the scatter plot using Seaborn
plt.figure(figsize=(8, 6))
scatter_plot = sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1', s=100)

# Step 4: Add labels for the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Step 5: Set the title of the plot
plt.title('Quadrant-wise Scatter Plot')

# Step 6: Add a legend
plt.legend(title='Quadrants')

# Display the plot
plt.show()
```

```python
'''18.With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.'''
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Generate data for the sine wave
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')

# Add the sine wave line
p.line(x, y, line_width=2)

# Add grid lines
p.grid.grid_line_alpha=0.3

# Display the plot in the notebook
output_notebook()
show(p)
```

```python
'''19.Using Bokeh, generate a bar chart of randomly generated categorical data, 
color bars based on their values, add hover tooltips to display exact values, 
label the axes, and set the title as 'Random Categorical Bar Chart'''

import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool, CategoricalColorMapper

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
data = {'category': np.random.choice(categories, size=10),
        'value': np.random.randint(1, 20, size=10)}
df = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a color mapper
color_mapper = CategoricalColorMapper(factors=categories,
                                      palette=['red', 'green', 'blue', 'orange', 'purple'])

# Create a figure
p = figure(title='Random Categorical Bar Chart',
           x_range=categories,
           x_axis_label='Category',
           y_axis_label='Value')

# Add the bar glyph
p.vbar(x='category', top='value', width=0.9, source=source,
       color={'field': 'category', 'transform': color_mapper})

# Add hover tooltips
hover = HoverTool(tooltips=[('Category', '@category'), ('Value', '@value')])
p.add_tools(hover)

# Display the plot in the notebook
output_notebook()
show(p)

```
