5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
# Generate a synthetic dataset with correlated features
np.random.seed(42)
size = 100  # number of samples

# Generate correlated features
feature_1 = np.random.normal(0, 1, size)
feature_2 = 0.8 * feature_1 + np.random.normal(0, 1, size)  # correlated with feature_1
feature_3 = 0.5 * feature_1 + 0.5 * feature_2 + np.random.normal(0, 1, size)  # correlated with both

# Create a DataFrame
df_corr = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3
})

# Compute the correlation matrix
corr_matrix = df_corr.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
days.

```python
# Generate the sales dataset
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data)

# Pivot the data to get the sales variation across different months and days
sales_pivot = df_sales.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Visualize the variation in sales across different months and days using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(sales_pivot, annot=True, cmap='Blues', linewidths=0.5, fmt='.0f')
plt.title('Heatmap of Sales Variation Across Months and Days')
plt.show()
```

5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create the bubble chart
plt.scatter(df['GDP'], df['Population'], s=df['Population'], alpha=0.5)  
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart of Population vs. GDP')

# Add country labels
for i, row in df.iterrows():
    plt.annotate(row['Country'], (row['GDP'], row['Population'])) 

plt.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Set output to display in the notebook (optional, if using Jupyter)
output_notebook()

# Generate x values from 0 to 10 and y values as sine of x
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the results
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource

# Set output to display in the notebook (optional, if using Jupyter)
output_notebook()

# Generate random x and y values
np.random.seed(42)  # Seed for reproducibility
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10

# Generate random sizes and colors
sizes = np.random.rand(100) * 30 + 10  # Marker sizes between 10 and 40
colors = np.random.rand(100) * 100     # Color range between 0 and 100

# Create a ColumnDataSource to hold the data
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a new plot
p = figure(title="Random Scatter Plot with Sizes and Colors", x_axis_label='x', y_axis_label='y')

# Create the scatter plot, using 'sizes' and 'colors' for marker size and color
mapper = linear_cmap(field_name='colors', palette='Viridis256', low=min(colors), high=max(colors))

p.circle('x', 'y', size='sizes', color=mapper, fill_alpha=0.6, source=source)

# Show the results
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Set output to display in the notebook (optional, if using Jupyter)
output_notebook()

# Given data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new figure
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', toolbar_location=None)

# Create a bar chart
p.vbar(x='fruits', top='counts', width=0.7, source=source, legend_field='fruits',
       fill_color=factor_cmap('fruits', palette=['#c9d9d3', '#718dbf', '#e84d60', '#ddb7b1'], factors=fruits))

# Customize the plot
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the results
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Set output to display in the notebook (optional, if using Jupyter)
output_notebook()

# Generate data and histogram
data_hist = np.random.randn(1000)  # Normally distributed data
hist, edges = np.histogram(data_hist, bins=30)  # Compute histogram

# Create a new figure for the histogram
p = figure(title="Histogram of Normally Distributed Data", 
           x_axis_label='Value', y_axis_label='Frequency')

# Plot the histogram using quad glyph
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color='navy', line_color='white', alpha=0.6)

# Show the plot
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import linear_cmap
from bokeh.layouts import layout
from bokeh.models import ColumnDataSource

# Set output to display in the notebook (optional, if using Jupyter)
output_notebook()

# Given data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flattening the data for Bokeh compatibility
source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    value=data_heatmap.flatten(),
))

# Define a color mapper with Viridis256 palette
color_mapper = LinearColorMapper(palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

# Create a new plot for the heatmap
p = figure(title="Bokeh Heatmap", x_axis_label='X', y_axis_label='Y',
           tools="hover", tooltips=[("Value", "@value")],
           x_range=(0, 1), y_range=(0, 1))

# Draw the heatmap as a set of rectangles
p.rect(x='x', y='y', width=1/10, height=1/10, source=source, 
       fill_color={'field': 'value', 'transform': color_mapper}, line_color=None)

# Add a color bar to the plot
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the results
show(p)
```
