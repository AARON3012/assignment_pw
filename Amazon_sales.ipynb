DATA CLEANING

```python
df.isnull().sum()
```

```python
df.rating_count
```

```python
df.rating_count = df.rating_count.str.replace(',','')
```

```python
df.rating_count = df.rating_count.astype(float)
```

```python
df.rating_count.fillna(df.rating_count.mean(),inplace=True)
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
df_copy = df.copy()
```

```python
df["discounted_price"] = df["discounted_price"].str.replace("₹", "").str.replace(",","")
```

```python
df["discounted_price"]= df["discounted_price"].astype(float)
```

```python
df.info()
```

```python
df.actual_price = df.actual_price.str.replace("₹", "").str.replace(",","")
```

```python
df.actual_price = df.actual_price.astype(float)
```

```python
df.discount_percentage = df.discount_percentage.str.replace("%","")
```

```python
df.discount_percentage = df.discount_percentage.astype(int)
```

```python
df.rating = df.rating.str.replace("|","")
```

```python
df.rating = df.rating.str.replace("out of 5 stars","")
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
print(df['rating'].dtype)
```

```python
df.info()
```

```python
#1.What is the average rating for each product category
df.groupby('category')['rating'].mean().sort_values(ascending=False)
```

```python
#2.What are the top rating_count products by category
df.groupby('category')['rating_count'].max().sort_values(ascending=False)
```

```python
#5.How does the average discount percentage vary across categories
df.groupby('category')['discount_percentage'].mean().sort_values(ascending=False)
```

```python
#5.What are the most popular product names
df.groupby('product_name')['rating_count'].max().sort_values(ascending=False)
```

```python
#6.What are the most popular product keywords
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer

# Make sure you've downloaded the stopwords if you haven't already
nltk.download('stopwords')

# Assuming your DataFrame is named df
# Combine product_name and about_product columns to extract keywords
text_data = df['product_name'].astype(str) + ' ' + df['about_product'].astype(str)

# Initialize a CountVectorizer to extract words (excluding stopwords)
vectorizer = CountVectorizer(stop_words='english', max_features=10)  # Adjust max_features as needed

# Fit and transform the text data
word_counts = vectorizer.fit_transform(text_data)

# Get the most common words
word_freq = pd.DataFrame(word_counts.toarray(), columns=vectorizer.get_feature_names_out())
word_freq_sum = word_freq.sum().sort_values(ascending=False)

# Display the most popular keywords
print(word_freq_sum.head(10))
```
