```python
#write a code that takes a list of words as input and returns a dictionary where the keys are unique words and the values are the frequencies of those words in the input list
def word_frequencies(words):
  """
  This function takes a list of words and returns a dictionary with word frequencies.
  """
  frequency = {}
  for word in words:
    frequency[word] = frequency.get(word, 0) + 1
  return frequency

# Example usage
words = ["apple", "banana", "apple", "cherry", "banana", "apple"]
result = word_frequencies(words)
print(result)  # Output: {'apple': 3, 'banana': 2, 'cherry': 1}
```

```python
#write a code that takes two dictionaries as input and merge them into single dictionary. if there are common keys,the value should be added together
def merge_dicts(dict1, dict2):
  merged = dict1.copy()
  for key, value in dict2.items():
    if key in merged:
      merged[key] += value
    else:
      merged[key] = value
  return merged

# Example usage
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 3, 'c': 4, 'd': 5}
merged_dict = merge_dicts(dict1, dict2)
print(merged_dict)  # Output: {'a': 1, 'b': 5, 'c': 7, 'd': 5}
```

```python
#write a code to access a value in a nested dictionary. the function should take the dictionary and a list of keys as input and return the corresponding value. if any keys do not exist in the dictionary,the function should return none.
def get_nested_value(nested_dict, key_list):
  """
  Accesses a value in a nested dictionary using a list of keys.

  Args:
    nested_dict: The nested dictionary.
    key_list: A list of keys to access the value.

  Returns:
    The corresponding value if all keys exist, otherwise None.
  """
  current_level = nested_dict
  for key in key_list:
    if key not in current_level:
      return None
    current_level = current_level[key]
  return current_level

 # Example usage
nested_dict = {"a": {"b": {"c": 10}}}
key_list = ["a", "b", "c"]
value = get_nested_value(nested_dict, key_list)
print(value)  # Output: 10

key_list = ["a", "b", "d"]
value = get_nested_value(nested_dict, key_list)
print(value)  #output: none
```

```python
#write a code that takes a dictionary as input and returns a sorteed version of it based on the values. you can choose whether to sort in ascending or descending order
def sort_dictionary_by_values(input_dict, ascending=True):
  """
  Sorts a dictionary by its values.

  Args:
    input_dict: The dictionary to sort.
    ascending: Whether to sort in ascending order (default) or descending order.

  Returns:
    A new dictionary sorted by values.
  """
  sorted_items = sorted(input_dict.items(), key=lambda item: item[1], reverse=not ascending)
  return dict(sorted_items)

# Example usage
my_dict = {'a': 3, 'b': 1, 'c': 2}
sorted_dict_asc = sort_dictionary_by_values(my_dict)
print(sorted_dict_asc)  # Output: {'b': 1, 'c': 2, 'a': 3}

sorted_dict_desc = sort_dictionary_by_values(my_dict, ascending=False)
print(sorted_dict_desc)  # Output: {'a': 3, 'c': 2, 'b': 1}
```

```python
#write a code that inverts a dictionary , swaapping keys and values. ensure that the inverted dictionary correctly handles cases where multiple keys have the same values by storing thr keys as a list in the inverted dictionary
from collections import defaultdict

def invert_dictionary(input_dict):
  """
  Inverts a dictionary, swapping keys and values.

  Args:
    input_dict: The dictionary to invert.

  Returns:
    A new dictionary with values as keys and keys as values (in a list if
    multiple keys have the same value).
  """
  inverted_dict = defaultdict(list)
  for key, value in input_dict.items():
    inverted_dict[value].append(key)
  return dict(inverted_dict)

# Example usage
my_dict = {'a': 1, 'b': 2, 'c': 1}
inverted_dict = invert_dictionary(my_dict)
print(inverted_dict)  # Output: {1: ['a', 'c'], 2: ['b']}
```
